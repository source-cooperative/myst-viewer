```python
import numpy as np
import xarray as xr

rng = np.random.default_rng(0)
temps = 15 + 8 * rng.standard_normal((3, 4))
da = xr.DataArray(
    temps,
    dims=("city", "month"),
    coords={
        "city": ["Lisbon", "Oslo", "Cairo"],
        "month": ["Jan", "Apr", "Jul", "Oct"],
    },
    name="temperature",
)
da
```

```python
# Mean temperature per city across the sampled months
da.mean(dim="month").round(2)
```

```python
# Label-based selection
print("Cairo in July:", float(da.sel(city="Cairo", month="Jul").round(2)))
```
