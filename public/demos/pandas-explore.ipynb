# pandas: tabular data exploration

A Jupyter notebook (`.ipynb`) demo using `pandas` and `numpy`, both bundled
with Pyodide (no install needed). DataFrames render as HTML tables in the
live output. Click **Activate**, then run each cell.

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(42)
df = pd.DataFrame({
    "city": ["Lisbon", "Oslo", "Cairo", "Tokyo", "Lima"],
    "population_m": [0.5, 0.7, 9.5, 14.0, 8.9],
    "avg_temp_c": 12 + 10 * rng.standard_normal(5),
})
df
```

```python
df.describe().round(2)
```

```python
df.sort_values("population_m", ascending=False).reset_index(drop=True)
```
